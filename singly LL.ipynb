```python
class Node:
    def __init__(self, element):
        self.element = element
        self.link = None
class Linked_List:
    def __init__(self):
        self.start = None

    def insert_at_beginning(self, element):
        new_node = Node(element)
        new_node.link = self.start
        self.start = new_node

    def insert_at_end(self, element):
        new_node = Node(element)
        if self.start is None:
            self.start = new_node
            return

        temp = self.start
        while temp.link:
            temp = temp.link

        temp.link = new_node

    def insert_at_middle(self, element, position):
        new_node = Node(element)

        if position == 1:
            new_node.link = self.start
            self.start = new_node
            return

        temp = self.start
        current_pos = 1

        while temp is not None and current_pos < position - 1:
            temp = temp.link
            current_pos += 1

        if temp is None:
            print("Position outside list range - inserting at end.")
            self.insert_at_end(element)
            return

        new_node.link = temp.link
        temp.link = new_node

    def search(self, key):
        temp = self.start

        while temp:
            if temp.element == key:
                print("Found")
                return True
            temp = temp.link

        print("Not Found")
        return False

    def delete_from_beginning(self):
        if self.start is None:
            print("List is Empty!")
            return
        self.start = self.start.link

    def delete_from_end(self):
        if self.start is None:
            print("List is Empty!")
            return

        if self.start.link is None:
            self.start = None
            return

        temp = self.start
        while temp.link.link:
            temp = temp.link

        temp.link = None

    def display(self):
        temp = self.start
        while temp:
            print(temp.element, end=" -> ")
            temp = temp.link
        print("None")

LL = Linked_List()

LL.insert_at_beginning(10)
LL.insert_at_beginning(20)
LL.insert_at_end(30)
LL.insert_at_end(40)
LL.display()

LL.insert_at_middle(50, 2)
LL.display()

LL.search(20)
LL.search(2)

LL.delete_from_beginning()
LL.display()

LL.delete_from_end()
LL.display()
```
