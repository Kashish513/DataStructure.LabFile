```python
class TicketingSystem:
    def __init__(self, limit):
        self.limit = limit
        self.queue = [None] * limit
        self.front = -1
        self.rear = -1

    def isFull(self):
        return self.rear == self.limit - 1

    def isEmpty(self):
        return self.front == -1 or self.front > self.rear

    # Add a new ticket
    def enqueue(self, ticket):
        if self.isFull():
            print("Queue Overflow! Cannot add:", ticket)
            return

        if self.front == -1:   # First ticket
            self.front = 0

        self.rear += 1
        self.queue[self.rear] = ticket
        print("Ticket inserted:", ticket)

    def dequeue(self):
        if self.isEmpty():
            print("Queue Underflow! No tickets to remove.")
            return

        removed = self.queue[self.front]
        print("Ticket processed:", removed)
        self.front += 1

        if self.front > self.rear:
            self.front = -1
            self.rear = -1

    def size(self):
        if self.isEmpty():
            return 0
        return (self.rear - self.front + 1)

    def display(self):
        if self.isEmpty():
            print("The queue has no tickets.")
        else:
            print("Tickets in Queue:", self.queue[self.front:self.rear + 1])

q = TicketingSystem(5)
q.enqueue("TK-01")
q.enqueue("TK-02")
q.enqueue("TK-03")
q.display()
q.dequeue()
q.display()
print("Total Tickets:", q.size())
print("Is Queue Full?", q.isFull())
print("Is Queue Empty?", q.isEmpty())
```
