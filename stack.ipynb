```python
history = []
forward_history = []
total_history = []

def insert_page(page):
    history.append(page)          
    forward_history.clear()      
    total_history.append(page)  

    print(f"Visited page: {page}")

def back_page():
    if len(history) < 2:
        print("No pages to go back to.")
        return

    last_page = history.pop()               
    forward_history.append(last_page)

    current = history[-1]                   
    total_history.append(current)          

    print(f"Going back from {last_page}")
    print(f"Current page: {current}")
def forward_page():
    if not forward_history:
        print("No forward pages available.")
        return
    next_page = forward_history.pop()       
    history.append(next_page)
    total_history.append(next_page)

    print(f"Going forward to: {next_page}")
def view_history():
    if not history:
        print("No pages in history.")
        return

    print("History:", " -> ".join(history))

def view_total_history():
    if not total_history:
        print("No total history.")
        return

    print("Total History:", " -> ".join(total_history))
def history_stats():
    if not total_history:
        print("No total history to analyze!")
        return

    print(f"\nTotal visits: {len(total_history)}")

    freq = {}
    for page in total_history:
        freq[page] = freq.get(page, 0) + 1

    print("\nPage Visit Frequency:")
    for page, count in freq.items():
        print(f"'{page}' visited {count} time(s)")

while True:
    print("\n1) Visit Page")
    print("2) Go Back")
    print("3) Go Forward")
    print("4) View History")
    print("5) View Total History")
    print("6) View History Stats")
    print("7) Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        page = input("Enter page name: ")
        insert_page(page)

    elif choice == 2:
        back_page()

    elif choice == 3:
        forward_page()

    elif choice == 4:
        view_history()

    elif choice == 5:
        view_total_history()

    elif choice == 6:
        history_stats()

    elif choice == 7:
        print("Exiting Browser... Goodbye!")
        break

    else:
        print("Invalid choice. Try again.")
```
